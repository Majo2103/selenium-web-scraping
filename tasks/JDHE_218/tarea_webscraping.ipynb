Los scrappers son herramientas poderosas para extraer automáticamente datos de una página web. ChatGPT puede ser una ayuda valiosa para generar expresiones XPath que indican dónde se encuentran los datos que deseamos extraer.

Debemos hacer el inspect y buscar un fragmento de html que le pueda dar contexto a ChatGTP sobre como hacer la consulta, Inspecciona la página hasta encontrar el nodo: ```<li class="col-xs-6 col-sm-4 col-md-3 col-lg-3">```, copia esta información y pide a chatGTP que te dé el XPATH de los títulos. Pruebalo haciendo ctrl+f en la herramienta de inspección

````python
#Obtener el XPATH del título de un libro en la página
titulos = []
elementos_libros = driver.find_elements(By.XPATH, '//li[@class="col-xs-6 col-sm-4 col-md-3 col-lg-3"]')
for elemento_libro in elementos_libros:
    titulo = elemento_libro.find_element(By.XPATH, './/h3/a').get_attribute('title')
    titulos.append(titulo)
    print(titulo)
````

Exploraremos cómo ChatGPT puede generar XPATH para extraer precios y otros datos específicos de una página web. 

````python
#Obtener el XPATH del URL de cada libro
urls = []
elementos_libros = driver.find_elements(By.XPATH, '//li[@class="col-xs-6 col-sm-4 col-md-3 col-lg-3"]')
for elemento_libro in elementos_libros:
    url = elemento_libro.find_element(By.XPATH, './/h3/a').get_attribute('href')
    urls.append(url)
    print(url)
````

````python
#Obtener el XPATH del URL de cada libro
precios = []
elementos_libros = driver.find_elements(By.XPATH, '//li[@class="col-xs-6 col-sm-4 col-md-3 col-lg-3"]')
for elemento_libro in elementos_libros:
    precio = elemento_libro.find_element(By.XPATH, './/p[@class="price_color"]').text
    precios.append(precio)
    print(precio)
````

Con la información obtenida, **intenta** guardarla en un DataFrame de pandas y encuentrar el libro más barato y el más caro de la primera página mostrada y brinda sus URL's

````python
df = pandas.DataFrame({'Title':titulos,'URL':urls,'Price':precios})
print(df)

# Encontrar el libro más barato
libro_mas_barato = df[df['Price'] == df['Price'].min()]

# Encontrar el libro más caro
libro_mas_caro = df[df['Price'] == df['Price'].max()]

# Mostrar los resultados
print(f"Libro más barato: {libro_mas_barato['Title'].values[0]} con url {libro_mas_barato['URL'].values[0]}")
print(f"Libro más caro: {libro_mas_caro['Title'].values[0]} con url {libro_mas_caro['URL'].values[0]}")
````

## Punto extra
Realiza la misma tarea que arriba pero considera todos los libros de la página

````python
elementos_libros = driver.find_elements(By.XPATH, '//li[@class="col-xs-6 col-sm-4 col-md-3 col-lg-3"]')
titulos = []
urls = []
precios = []
# Iterar sobre los elementos y extraer información
while True:
    # Iterar sobre los elementos y extraer información
    for elemento_libro in elementos_libros:
        titulo = elemento_libro.find_element(By.XPATH, './/h3/a').get_attribute('title')
        url = elemento_libro.find_element(By.XPATH, './/h3/a').get_attribute('href')
        precio = elemento_libro.find_element(By.XPATH, './/p[@class="price_color"]').text
        disponibilidad = elemento_libro.find_element(By.XPATH, './/p[@class="instock availability"]').text
        print(f"Título: {titulo}")
        print(f"URL: {url}")
        print(f"Precio: {precio}")
        print("\n")
        titulos.append(titulo)
        urls.append(url)
        precios.append(precio)

    # Verificar si hay un botón "next"
    try:
        boton_next = driver.find_element(By.XPATH, '//li[@class="next"]/a')
        # Si hay un botón "next", hacer clic y avanzar a la siguiente página
        boton_next.click()
        # Actualizar la lista de elementos_libros para la nueva página
        elementos_libros = driver.find_elements(By.XPATH, '//li[@class="col-xs-6 col-sm-4 col-md-3 col-lg-3"]')
    except:
        # Si no hay un botón "next", salir del bucle
        break
    
df_ext = pandas.DataFrame({'Title':titulos,'URL':urls,'Price':precios})
df_ext

# Encontrar el libro más barato
libro_mas_barato = df_ext[df_ext['Price'] == df_ext['Price'].min()]

# Encontrar el libro más caro
libro_mas_caro = df_ext[df_ext['Price'] == df_ext['Price'].max()]

# Mostrar los resultados
print(f"Libro más barato: {libro_mas_barato['Title'].values[0]} con url {libro_mas_barato['URL'].values[0]}")
print(f"Libro más caro: {libro_mas_caro['Title'].values[0]} con url {libro_mas_caro['URL'].values[0]}")
````
